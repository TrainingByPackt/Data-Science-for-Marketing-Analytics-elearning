```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
def sigmoid(x):
  return 1/(1+np.exp(-x))
```

```python
z=np.arange(-10,10,0.1)
z
```

```python
sig=sigmoid(z)
```

```python
plt.plot(z,sig)
plt.yticks([0.0,0.5,1.0])
plt.grid(True)
```

```python
plt.plot(z,0+0.5*z)
plt.grid(True)
```
